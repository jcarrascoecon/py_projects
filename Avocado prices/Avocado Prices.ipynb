```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('avocado.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
region = df.groupby('region')
means = region.mean()
means = means.sort_values(by = 'AveragePrice', ascending = False)
means
```

```python
volume = region.sum()
volume.sort_values(by = 'Total Volume', ascending = False)
```
