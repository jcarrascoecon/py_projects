```python
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python

df = pd.read_csv("Foreign_Exchange_Rates.csv")

df['Time Serie'] = pd.to_datetime(df['Time Serie'])

df = df.drop('Unnamed: 0', axis = 1)

df = df.rename(columns = {
    'AUSTRALIA - AUSTRALIAN DOLLAR/US$':'AUS/USD',
    'EURO AREA - EURO/US$' : 'EU/USD',
    'UNITED KINGDOM - UNITED KINGDOM POUND/US$': 'GBP/USD',
    'CHINA - YUAN/US$' : 'RMB/USD',
})

df['AUS/USD'] = pd.to_numeric(df['AUS/USD'], errors = 'coerce')

df = df.dropna()

df.head(20)

df = df.set_index('Time Serie')
```

```python
df['2015']
```

```python
small = df['AUS/USD'].resample('W').max()
small.plot()
```

```python
Max_Weekly = df['AUS/USD'].resample('W').max()
Min_Weekly = df['AUS/USD'].resample('W').min()
```

```python
plt.plot(Max_Weekly)
plt.plot(Min_Weekly)
plt.legend(loc='lower left')
plt.xlabel('Year')
plt.ylabel('AUS/USD')
plt.show()
```

```python
weekly_band = Max_Weekly-Min_Weekly

plt.plot(weekly_band)
```

```python
big_spike = df['2008':'2012']
big_spike
max_spike = big_spike['AUS/USD'].resample('W').max()
min_spike = big_spike['AUS/USD'].resample('W').min()
plt.plot(max_spike)
plt.plot(min_spike)
```
