```python
import numpy as np
import matplotlib as plt

r = 0.025
T = 50
b = np.empty(T+1)
b[0] = 1000

for t in range(T):
    b[t+1] = b[t] * (1 + r)

plt.plot(b, label = "value")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
alphas = [0,0.8,0.98]


for alpha in alphas:
    x = np.empty(200)
    x[0] = 0
    for i in range(T-1):
        xtplus = alpha * x[i] + np.random.randn()
        x[i+1] = xtplus
    plt.plot(x)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

T = 200
alpha = 0.9
x = np.empty(200)
x[0] = 0

for i in range(T-1):
    if x[i] > 0:
        xtplus = alpha * x[i] + np.random.randn()
    else:
        xtplus = -1 * alpha * x[i] + np.random.randn()
    x[i+1] = xtplus

plt.plot(x)
plt.show()
```

```python
n = 100000

count = 0
for i in range(n):
    u, v = np.random.uniform(), np.random.uniform()
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)
    if d < 0.5:
        count += 1

area_estimate = count / n

print(area_estimate * 4)
```
