```python
import pandas as pd
import numpy as np
import qeds

df = pd.DataFrame({"numbers": ["#23", "#24", "#18", "#14", "#12", "#10", "#35"],
                   "nums": ["23", "24", "18", "14", np.nan, "XYZ", "35"],
                   "colors": ["green", "red", "yellow", "orange", "purple", "blue", "pink"],
                   "other_column": [0, 1, 0, 2, 1, 0, 2]})
df

c2n = "#39"
c2n = int(c2n.replace("#", ""))
print(c2n)

df["numbers_str"] = df["numbers"].str.replace("#", "")
df["numbers_numeric"] = pd.to_numeric(df["numbers_str"])
df["colors_upper"] = df['colors'].str.upper()
df.dtypes
df['nums_tonumeric'] = pd.to_numeric(df['nums'], errors = 'coerce')
df.dropna(inplace=True)
df
```

```python
chip = qeds.data.load("chipotle_raw")
chip.head()
chip.dtypes
chip['item_price'] = chip['item_price'].str.replace("$","")
chip['item_price'] = pd.to_numeric(chip['item_price'])
chip.head()
```

```python

chicken_chip = chip.loc[chip["item_name"].isin(['Chicken Bowl','Chicken Burrito','Chicken Tacos', 'Chicken Salad'])]
print(chicken_chip['item_price'].mean())

steak_chip = chip.loc[chip["item_name"].isin(['Steak Bowl','Steak Burrito','Steak Tacos','Steak Salad'])]
print(steak_chip['item_price'].mean())

print(steak_chip['item_price'].sum())
print(chicken_chip['item_price'].sum())

```

```python
chips = qeds.data.load("chipotle_raw")
def missing_items(i):
    x = i.dropna()
    n = []
    for p in list(i.columns):
        n.append(len(i[p])-len(x[p]))
    return n
print(missing_items(chip))

def missing_columns(d):
    n = []
    for p in list(d.columns):
        x = d[p]
        y = x.dropna()
        n.append(len(x)-len(y))
    return n

print(missing_columns(chips))
```
