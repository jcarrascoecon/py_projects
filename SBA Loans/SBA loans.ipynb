```python
dict = {
    'a $5-10 million': 5,
    'b $2-5 million': 2,
    'c $1-2 million': 1,
    'd $350,000-1 million': 0.35,
    'e $150,000-350,000': 0.15
}
df['LoanRange'] = df['LoanRange'].map(dict)
```

```python
df.head()
```

```python
def most_aid(x):
    arg = df.groupby(x)
    arg_sum = arg.sum()
    result = arg_sum['LoanRange'].max()
    print(arg_sum.loc[arg_sum['LoanRange'] == result])
    return result

print(most_aid('State'))
```

```python
print(most_aid('City'))
```

```python
print(most_aid('Zip'))
```

```python
def most_jobs(x):
    arg = df.groupby(x)
    arg_sum = arg.sum()
    result = arg_sum['JobsRetained'].max()
    print(arg_sum.loc[arg_sum['JobsRetained'] == result])
    return result
print(most_jobs('State'))
print(most_jobs('City'))
print(most_jobs('Zip'))
```

```python
arg = df.groupby('State')
arg_sum = arg.sum()
sorted_arg = arg_sum.sort_values('LoanRange', ascending = False)
x = sorted_arg.drop('Zip', axis = 1)
x = x.drop('JobsRetained', axis = 1)
x = x.drop('NAICSCode', axis = 1)
x = x.rename(columns = {'LoanRange':'Ordinal Ranking of Loans'})
pd.options.display.float_format = '{:,}'.format
x
```
